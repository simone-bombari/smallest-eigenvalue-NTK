```python
import torchvision
import torch
from torch.utils.data import DataLoader
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from utils import loss_calculator, compute_loss_accuracy, build_Gaussian_dataset, call_partial_mnist
from utils import FullyConnected_4relu, compute_ntk, FullyConnected_4relu_mnist



N = 10
n = 100
d = 5
device = 'cpu'

dataset_train = call_partial_mnist(N, N)
dataset_train_g = build_Gaussian_dataset(N, d)

batch_size = N
lr = 1 / (10 * np.sqrt(N))

train_loader = torch.utils.data.DataLoader(dataset_train, batch_size=batch_size)
train_loader_g = torch.utils.data.DataLoader(dataset_train_g, batch_size=batch_size)


stuff = next(iter(train_loader))
stuff_g = next(iter(train_loader_g))
```

```python
net = FullyConnected_4relu_mnist(n).to(device)
loss_function = 'MSE'
optimizer = optim.Adam(net.parameters(), lr=lr)
epochs = 100
previous_loss = np.inf

for epoch in range(epochs):

    if epoch % 100 == 0:
        if epoch > 0:
            if average_loss > previous_loss:
                break
            else:
                previous_loss = average_loss
        average_loss = 0

    # total_labels = np.zeros(10)
    for input_images, labels in iter(train_loader):
        input_images, labels = input_images.to(device), labels.to(device).type(torch.FloatTensor)
        optimizer.zero_grad()
        outputs = net(input_images)
        loss = loss_calculator(outputs, labels, loss_function, None, device)
        loss.backward()
        optimizer.step()

    print('Epoch {}:\nTrain loss = {:.7f}\n'.format(
           epoch+1, loss), flush=True)

    average_loss += loss

previous_loss = previous_loss.item() / 50
```
